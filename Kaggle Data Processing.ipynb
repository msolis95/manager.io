# Outfield Players

```python
# downloading the data
outfield_players = players[players['position'] != 'Goalkeeper'].reset_index()
outfield_players = outfield_players[["player_id", "position", "height_in_cm", "city_of_birth", "market_value_in_eur", "last_season", "current_club_id"]]
outfield_players = outfield_players.rename(columns={"height_in_cm": "goals"})
outfield_players = outfield_players.rename(columns={"city_of_birth": "assists"})
outfield_players = outfield_players.rename(columns={"market_value_in_eur": "passing_accuracy"})
outfield_players = outfield_players.rename(columns={"last_season": "key_passes_per_game"})
outfield_players = outfield_players.rename(columns={"current_club_id": "interceptions_per_game"})

# randomizing stats
num_rows = outfield_players.shape[0]

max_goals = 60

outfield_players["goals"] = [random.randint(0, max_goals) for i in range(num_rows)]

max_assists = 30
outfield_players["assists"] = [random.randint(0, max_assists) for i in range(num_rows)]

pass_accur_max = 100
pass_accur_lst = []
for i in range(num_rows):
  random_key_passes = random.uniform(0, pass_accur_max)
  random_key_passes = round(random_key_passes, 2)
  pass_accur_lst.append(random_key_passes)

outfield_players["passing_accuracy"] = [pass_accur_lst[i] for i in range(num_rows)]

max_key_passes = 7
max_key_passes_lst = []
for i in range(num_rows):
  random_key_passes = random.uniform(0, max_key_passes)
  random_key_passes = round(random_key_passes, 2)
  max_key_passes_lst.append(random_key_passes)

outfield_players["key_passes_per_game"] = [max_key_passes_lst[i] for i in range(num_rows)]

interceptions_max = 10
interceptions_lst = []
for i in range(num_rows):
  random_key_passes = random.uniform(0, interceptions_max)
  random_key_passes = round(random_key_passes, 2)
  interceptions_lst.append(random_key_passes)

outfield_players["interceptions_per_game"] = [interceptions_lst[i] for i in range(num_rows)]

# fixing positions
pos_map = {"Defender" : "D", "Midfield" : "M", "Attack" : "F"}
for i in range(num_rows):
  pos = outfield_players.at[i, "position"]
  outfield_players.at[i, "position"] = pos_map.get(pos, random.choice(list(pos_map.values())))

display(outfield_players)
```

# Goalkeepers

```python
goalkeepers = players[players['position'] == 'Goalkeeper'].reset_index()
goalkeepers = goalkeepers[["player_id", "last_season", "current_club_id"]]
goalkeepers = goalkeepers.rename(columns={"last_season": "saves"})
goalkeepers = goalkeepers.rename(columns={"current_club_id": "clean_sheets"})
max_clean_sheets = 25
max_clean_sheets_lst = []
num_rows = goalkeepers.shape[0]
saves_lst = []

for i in range(max_clean_sheets):
  max_clean_sheets_lst.append(i + 1)

for i in range(180):
  saves_lst.append(i + 1)

goalkeepers["saves"] = [random.choice(saves_lst) for _ in range(num_rows)]
goalkeepers["clean_sheets"] = [random.choice(max_clean_sheets_lst) for _ in range(num_rows)]
display(goalkeepers)
```

# Clubs

```python
clubs = clubs.rename(columns={"squad_size": "club_country"})
clubs = clubs.rename(columns={"name": "club_name"})
clubs = clubs.drop("club_code", axis=1)
clubs = clubs.drop("domestic_competition_id", axis=1)
clubs = clubs.drop("total_market_value", axis=1)
clubs = clubs.drop("average_age", axis=1)
clubs = clubs.drop("foreigners_number", axis=1)
clubs = clubs.drop("foreigners_percentage", axis=1)
clubs = clubs.drop("national_team_players", axis=1)

clubs = clubs.drop("stadium_name", axis=1)
clubs = clubs.drop("stadium_seats", axis=1)
clubs = clubs.drop("net_transfer_record", axis=1)
clubs = clubs.drop("coach_name", axis=1)
clubs = clubs.drop("last_season", axis=1)
clubs = clubs.drop("filename", axis=1)
clubs = clubs.drop("url", axis=1)

num_rows = clubs.shape[0]
lst_countries = competitions['country_name'].tolist()
clubs["club_country"] = [random.choice(lst_countries) for _ in range(num_rows)]
display(clubs)
```

# Players



```python
def random_birth_date():
  birth_date = ""
  random_year = random.randint(1980, 2000)
  random_month = random.randint(1, 12)
  if random_month == 2:
    random_day = random.randint(1, 28)

  elif random_month in [1, 3, 5, 7, 8, 10, 12]:
    random_day = random.randint(1, 31)

  else:
    random_day = random.randint(1, 30)

  birth_date += str(random_year)
  if len(str(random_month)) == 1:
    birth_date += "-0"
  else:
    birth_date += "-"

  birth_date += str(random_month)

  if len(str(random_day)) == 1:
    birth_date += "-0"
  else:
    birth_date += "-"

  birth_date += str(random_day)
  return birth_date

def random_nationality():
  nats = players['nationality'].tolist()
  nat = random.choice(nats)
  while nat != nat:
    nat = random.choice(nats)
  return nat
```

```python
players = players.rename(columns={"name": "player_name"})
players = players.rename(columns={"current_club_id": "club_id"})
players = players.rename(columns={"date_of_birth": "birth_date"})
players = players.rename(columns={"country_of_citizenship": "nationality"})
players = players.rename(columns={"market_value_in_eur": "market_value"})
players = players.rename(columns={"foot": "num_appearances"})
players = players[["player_id", "player_name", "club_id", "birth_date", "nationality", "market_value", "num_appearances"]]

num_rows = players.shape[0]
max_appearances = 75
players["num_appearances"] = [random.randint(0, max_appearances) for i in range(num_rows)]

min_val = 2000000
max_val = 100000000

# Replace NaN values in "column_name" with random integers between min_val and max_val
players['market_value'] = players['market_value'].apply(lambda x: np.random.randint(min_val, max_val) if np.isnan(x) else x)

players['birth_date'] = players['birth_date'].apply(lambda x: random_birth_date() if x != x else x)

players['nationality'] = players['nationality'].apply(lambda x: random_nationality() if x != x else x)

display(players)
```

```python
num_rows = len(players)
max_len = 0
nationality = ""
for i in range(num_rows):
  if type(players.at[i, "nationality"]) != str:
    print(i)
  nat_len = len(players.at[i, "nationality"])
  if nat_len > max_len:
    max_len = nat_len
    nationality = players.at[i, "nationality"]
print(max_len)
print(nationality)
```
